# Hipotesis Testing 

#H0 = There is a significant different of cost of Video advertisement and Text in daily basis for advertisement 

```python
t_stat, p_val = stats.ttest_ind(daily_video, daily_image)
print('P-value:',p_val[0]) #the p-value isn't divided by 2 since the output is two-sided p-value
print('t-statistics:',t_stat[0])
```

```python
video_pop = np.random.normal(daily_video.spend_usd.mean(),daily_video.spend_usd.std(),10000)
text_pop = np.random.normal(daily_text.spend_usd.mean(),daily_text.spend_usd.std(),10000)

ci = stats.norm.interval(0.95, daily_video.spend_usd.mean(), daily_video.spend_usd.std())
plt.figure(figsize=(15,15))
sns.distplot(video_pop, label='Average VIDEO Cost a Day',color='green')
sns.distplot(text_pop, label='Average TEXT Cost a Day',color='blue')

plt.axvline(daily_video.spend_usd.mean(), color='blue', linewidth=2, label='VIDEO Mean')
plt.axvline(daily_text.spend_usd.mean(), color='red',  linewidth=2, label='TEXT Mean')

plt.axvline(ci[1], color='green', linestyle='dashed', linewidth=2, label='Confidence Threshold of 95%')
plt.axvline(ci[0], color='green', linestyle='dashed', linewidth=2)

plt.axvline(video_pop.mean()+t_stat[0]*video_pop.std(), color='black', linestyle='dashed', linewidth=2, label = 'Alternative Hypothesis')
plt.axvline(video_pop.mean()-t_stat[0]*video_pop.std(), color='black', linestyle='dashed', linewidth=2)

plt.legend()
```

#Hypotesis is fail to reject because there was not a signficant different between text and video in daily cost
Also it can be found that the average cost of Text and Video advertisements are very similar
